# Renewable Energy Impact on Spending

```python
df1.energy_bin.unique()
```

```python
seds_re = df1[df1["energy_bin"] == "Renewable"]
seds_re = seds_re[seds_re["unit"] == "Million kilowatthours"]
seds_re_us = seds_re[seds_re["state_code"] == "US"]
seds_re_us_kwh = seds_re_us[seds_re_us["unit"] == "Million kilowatthours"]
seds_re_us_kwh
```

```python
#draw plot of US RE consumption
total_consump = seds_re_us_kwh.groupby("year").sum()
plt.scatter(total_consump.index, total_consump.value)
```

```python
#RE consumption in states that don't focus on manufacturing
non_manufacturing_states = ["OK", "IA", "OH", "DE", "IL", "UT", "OR", "HI", 
                            "NJ", "WV", "ND", "WY", "GA", "MD", "AZ", "GU", "VT"]
green_states_re = seds_re[seds_re["state_code"].isin(non_manufacturing_states)]
not_green_states_re = seds_re[~(seds_re["state_code"].isin(non_manufacturing_states))]

green_states_re.head()
```

```python
#Group by year
green_consump = green_states_re.groupby("year").sum()
plt.plot(green_consump.index, green_consump.value)
```

```python
not_green_consump = not_green_states_re.groupby("year").sum()
plt.plot(not_green_consump.index, not_green_consump.value)
```

Consumption seems pretty similiar between states that focus on green initiatives and those that don't. Let's compare how manufacturing spending compares in these states.

```python
non_renew = ['Petroleum','Natural Gas', 'Coal']
seds_manu = df1[df1["energy_bin"].isin(non_renew)]
seds_manu_green_states = seds_manu[seds_manu["state_code"].isin(non_manufacturing_states)]
seds_manu_not_green_states = seds_manu[~(seds_manu["state_code"].isin(non_manufacturing_states))]
```

```python
green_state_manu_consump = seds_manu_green_states.groupby("year").sum()
plt.plot(green_state_manu_consump.index, green_state_manu_consump.value)
```

```python
not_green_state_manu_consump = seds_manu_not_green_states.groupby("year").sum()
plt.plot(not_green_state_manu_consump.index, not_green_state_manu_consump.value)
```
